```python
# Data preprocessing
```

```python
# Normalizing the  column 'Amount' feature to ensure all the features contribute equally to the model.
from sklearn.preprocessing import StandardScaler
df['normalizedAmount'] = StandardScaler().fit_transform(df['Amount'].values.reshape(-1,1))
data = df.drop(['Amount'], axis=1)
```

```python
#  Rearranging the columns Move 'NormalizedAmount' column next to other features
cols = data.columns.tolist()
cols = cols[:-2] + ['normalizedAmount'] + ['Class']
data = data[cols]
```

```python
# Separate features (X) and target variable (y)
X = data.drop('Class', axis=1)
y = data['Class']
```

```python


from imblearn.over_sampling import SMOTE

# Initializing SMOTE
sm = SMOTE(random_state=42)

# Before resampling, removing the  rows with NaN in 'Class' column
data = data.dropna(subset=['Class'])

# Separate features (X) and target variable (y) after removing NaN
X = data.drop('Class', axis=1)
y = data['Class']

# Resampling the dataset
X_resampled, y_resampled = sm.fit_resample(X, y)

# Displaying new class distribution
print("Original dataset shape:", y.value_counts())
print("Resampled dataset shape:", y_resampled.value_counts())
```

```python
# model development
```

```python
from sklearn.model_selection import train_test_split

# Split the resampled data
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.3, random_state=42
)
```

```python
# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier

# Initialize the Random Forest Classifier Algorithm
random_forest = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
random_forest.fit(X_train, y_train)
```

```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier

# Initialize the Decision Tree Classifier
decision_tree = DecisionTreeClassifier(random_state=42)

# Train the model
decision_tree.fit(X_train, y_train)
```

```python
# Model Evaluation
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, classification_report

def evaluate_model(model, X_test, y_test):
    # Predict the labels
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]

    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred_proba)

    # Print classification report
    print(classification_report(y_test, y_pred))

    # Return the metrics
    return accuracy, precision, recall, f1, roc_auc
```

```python
# Evaluation of Random Forest classifier
print("Random Forest Evaluation Metrics:")
rf_metrics = evaluate_model(random_forest, X_test, y_test)
```

```python
# Evaluation of Decision Tree classifier
print("Decision Tree Evaluation Metrics:")
dt_metrics = evaluate_model(decision_tree, X_test, y_test)
```

```python
# Confusion Metrices
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(cm, classes, title='Confusion Matrix', cmap=plt.cm.Blues):
    """
    Plots the confusion matrix.
    """
    plt.figure(figsize=(5,5))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    # Print numbers inside the squares
    thresh = cm.max() / 2
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(
            j, i, format(cm[i, j], 'd'),
            horizontalalignment='center',
            color='white' if cm[i, j] > thresh else 'black'
        )

    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.show()
```

```python
# Random Forest Confusion Matrix
import itertools

# Compute confusion matrix
cm_rf = confusion_matrix(y_test, random_forest.predict(X_test))

# Plot confusion matrix
plot_confusion_matrix(cm_rf, classes=[0, 1], title='Random Forest Confusion Matrix')
```

```python
# Decison Tree Confusion Matrix
cm_dt = confusion_matrix(y_test, decision_tree.predict(X_test))
plot_confusion_matrix(cm_dt, classes=[0, 1], title='Decision Tree Confusion Matrix')
```

```python
# ROC Curves
from sklearn.metrics import RocCurveDisplay

# Plot ROC curve for Random Forest
RocCurveDisplay.from_estimator(random_forest, X_test, y_test)
plt.title('Random Forest ROC Curve')
plt.show()
```

```python
# Plot ROC curve for Decision Tree
RocCurveDisplay.from_estimator(decision_tree, X_test, y_test)
plt.title('Decision Tree ROC Curve')
plt.show()
```

```python
# Modal Results Comparison
# Creating a DataFrame to compare models
result = pd.DataFrame({
    'Model': ['Random Forest', 'Decision Tree'],
    'Accuracy': [rf_metrics[0], dt_metrics[0]],
    'Precision': [rf_metrics[1], dt_metrics[1]],
    'Recall': [rf_metrics[2], dt_metrics[2]],
    'F1 Score': [rf_metrics[3], dt_metrics[3]],
    'ROC AUC': [rf_metrics[4], dt_metrics[4]]
})

# Display the results
print(result)
```
